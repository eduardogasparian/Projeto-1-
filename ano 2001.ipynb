```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dados_todos = pd.read_excel('todos_dados_comparacoes.xlsx')
dados_idh = pd.read_excel('hdi_human_development_index.xlsx')
dados_cm = pd.read_excel('child_mortality_0_5_year_olds_dying_per_1000_born.xlsx')
dados_unem = pd.read_excel('long_term_unemployment_rate_percent.xlsx')
```

```python
dados_idh.head(5)
di = dados_idh.set_index('geo')
di.head(5)
```

```python
dados_unem.head(5)
du = dados_unem.set_index('geo')
du.head(5)
```

```python
dados_cm.head(5)
dc = dados_cm.set_index('geo')
dc.head(5)
```

```python
idh2001 = di.loc[:,[2001]]
idh2001.head(5)
```

```python
u2001 = du.loc[:,[2001]]
u2001.head(5)
```

```python
c2001 = dc.loc[:,[2003]]
c2001.head(5)
```

```python
c2001 = dc.loc[:,[2001]].add_suffix('_cm')
c2001.head(5)
```

```python
idh_u = idh2001.join(u2001,how = 'inner',lsuffix = '_idh',rsuffix = '_unem')
idh_u.head(5)
```

```python
final_2 = idh_u.join(c2001,how='inner')
final_2.head(2)
```

```python
final_2.plot.scatter(x = '2001_unem',y = '2001_idh');
sns.regplot(final_2['2001_unem'],final_2['2001_idh'], color = ('blue'))
plt.grid()
```

```python
final_2.plot.scatter(x = '2001_cm',y = '2001_idh');
sns.regplot(final_2['2001_cm'],final_2['2001_idh'], color = ('blue'))
plt.grid()
```

```python
final_2.to_excel( 'todos_dados_comparacoes ano 2001.xlsx', index=True)
```
